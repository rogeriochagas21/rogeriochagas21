```python
df_Fim = df.loc[df["Final de Semana"] == "Final de Semana"]
df_Semana = df.loc[df["Final de Semana"] == "Semana"]

df_Fim['Mes'] = df['Data'].dt.strftime('%Y-%m') 
df_Fim['Mes']
df_mes_Fim = df.groupby(df_Fim['Mes'], as_index=False).mean() 
#df_mes_Fim.round(2)

df_Semana['Mes'] = df['Data'].dt.strftime('%Y-%m') 
df_Semana['Mes']
df_mes_Semana = df.groupby(df_Semana['Mes'], as_index=False).mean() 
df_mes_Semana.round(2)

df1 = df_mes_Fim["Consumo de cerveja (litros)"]
df2 = df_mes_Semana["Consumo de cerveja (litros)"]

dfFinal = [df1, df2]
dfFinal
```

```python
# A função chi2_contingency() do módulo scipy.stats toma como entrada a tabela de contingência 
# no formato de array 2d. Ele retorna uma tupla contendo estatísticas de teste, o valor p, 
# graus de liberdade e a tabela esperada (criada a partir dos valores calculados), nessa ordem. 
from scipy.stats import chi2_contingency

score = dfFinal
stat, p, dof, expected = chi2_contingency(score) #Infos teóricas sobre cada parâmetro (stat, p...) no primeiro link "Teo sobre Qui-Qua"
```

```python
alpha = 0.05

print("O valor resultante do teste é: " + str(p))
if p <= alpha:
    print('Dependente (rejeita H0).')
else:
    print('Independente (não rejeita H0).')
```

```python
#Fazer amplitude dos dados e depois fazer a contagem dos dados que estão entre as faixas.
#Aula 15/08. 

max = df["Consumo de cerveja (litros)"].max()
min = df["Consumo de cerveja (litros)"].min()

faixas = 15

amplitude = (max-min)/faixas

min
```

```python
#Calculando vetor pro fim de semana

max = df["Consumo de cerveja (litros)"].max()
min = df["Consumo de cerveja (litros)"].min()
valores_Fim = []
faixa = min
while(faixa < max):
  minaux = faixa
  faixa = faixa + amplitude
  maxaux = faixa
  concfi = (df_Fim["Consumo de cerveja (litros)"]>=minaux)
  concfi2 = (df_Fim["Consumo de cerveja (litros)"]<= maxaux)

  aux = df_Fim.loc[(concfi & concfi2),"Consumo de cerveja (litros)"].count()
  valores_Fim.append(aux)

valores_Fim
```

```python
#Calculando vetor para Semana

max = df["Consumo de cerveja (litros)"].max()
min = df["Consumo de cerveja (litros)"].min()
valores_Semana = []
faixa = min
while(faixa < max):
  minaux = faixa
  faixa = faixa + amplitude
  maxaux = faixa
  concfi = (df_Semana["Consumo de cerveja (litros)"]>=minaux)
  concfi2 = (df_Semana["Consumo de cerveja (litros)"]<= maxaux)

  aux = df_Semana.loc[(concfi & concfi2),"Consumo de cerveja (litros)"].count()
  valores_Semana.append(aux)

valores_Semana
```

```python
from scipy.stats import chi2_contingency

score = [valores_Fim, valores_Semana]
stat, p, dof, expected = chi2_contingency(score) #Infos teóricas sobre cada parâmetro (stat, p...) no primeiro link "Teo sobre Qui-Qua"
```

```python
alpha = 0.05

print("O valor resultante do teste é: " + str(p))
if p <= alpha:
    print('Dependente (rejeita H0).')
else:
    print('Independente (não rejeita H0).')
```

```python
import numpy as np
from scipy.stats import pearsonr
```

```python
# Função para calcular o coeficiente de correlação de Pearson entre dois arrays
def corr_coef(x,y,label=None,color=None,**kwargs):
    
    # Calcula o valor
    r,p = pearsonr(x,y)
    
    # Adiciona o rótulo ao gráfico 
    ax = plt.gca()
    ax.annotate('r = {:.2f}'.format(r), xy=(0.5,0.5), xycoords='axes fraction', ha='center')
    ax.set_axis_off()
```

```python
g = sns.PairGrid(df);

g.map_diag(sns.distplot);
g.map_lower(sns.scatterplot);
g.map_upper(corr_coef);
```
