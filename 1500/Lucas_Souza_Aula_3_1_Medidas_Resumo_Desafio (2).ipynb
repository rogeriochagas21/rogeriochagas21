```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid", #rc=custom_params, 
              palette="colorblind", rc={'figure.figsize':(7,6)})


df = pd.read_excel("/content/Lucas Souza - Análise Descritiva (Medidas).xlsx")
df.loc[df["region"] == "INNER_CITY","income"].sum()

df
```

```python
#Gráfico de barras: Salário x Região
sns.barplot(
    x="income", 
    y="region",
    data=df, 
    estimator=sum, 
    ci=None, 
    palette = ["#414FB1","#65B3C7", "green", "#DA556D"],
    saturation = 2
);
```

```python
#Gráfico de barras: Salário x Região
sns.barplot(
    x="income", 
    y="region", 
    hue="married",
    data=df, 
    estimator=sum, 
    ci=None, 
    saturation = 2
);
```

```python
#Gráfico de barras: Salário x Região
sns.barplot(
    x="income", 
    y="region", 
    hue="children",
    data=df, 
    estimator=sum, 
    ci=None, 
    saturation = 2
);
```

```python
sns.lineplot(x="age", y="income",
             hue="sex", 
             #style="married", # ou size
             ci= None,
             estimator = sum,
             #markers=True, 
             #dashes=False,
             data=df);
```

```python
dfMALE = df.loc[df["sex"] == "MALE"]

sns.regplot(
    data=dfMALE, x="age", y="income",
);
```
