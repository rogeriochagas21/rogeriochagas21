```python
df = pd.read_csv("/content/Dados - Desafio 2.csv", sep=";",
     parse_dates = ["dteday"], decimal = ","
)
```

```python
df.dtypes
```

```python
df
```

```python
pd.isna(df["season"]).sum
```

```python
df.loc[df["weathersit"] == "Clear, Few clouds, Partly cloudy, Partly cloudy", "season"] = "springer"
df.loc[df["weathersit"] == "Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist", "season"] = "summer"
df.loc[df["weathersit"] == "Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog", "season"] = "winter"
df.loc[df["weathersit"] == "Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds", "season"] = "fall"
```

```python
df
```

```python
pd.isna(df["season"]).sum
```

```python
df.isnull().sum()
```

```python
df['temp'].fillna(df['temp'].mean(), inplace = True)
df['atemp'].fillna(df['atemp'].mean(), inplace = True)
df['hum'].fillna(df['hum'].mean(), inplace = True)
df['windspeed'].fillna(df['windspeed'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace = True, how = 'any')
```
